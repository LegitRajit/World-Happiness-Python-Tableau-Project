# 03 Exploring Relationships

```python
# This option ensures the charts you create are displayed in the notebook without the need to "call" them specifically.

%matplotlib inline
```

```python
df_whr22.head()
```

```python
df_whr_M.head()
```

```python
# Creating Subsets of Data (Removing Year and Country columns)

sub22 = df_whr22[['Happiness Score','GDP per Capita','Social Support','Healthy Life Expectancy','Freedom to make life choices','Generosity', 'Perceptions of Corruption']]

# Creating Same subset for merged dataset
subM = df_whr_M[['Happiness Score','GDP per Capita','Social Support','Healthy Life Expectancy','Freedom to make life choices','Generosity', 'Perceptions of Corruption']]
```

## Creating a Correlation Matrix

```python
# Create a correlation matrix using pandas (2022)

sub22.corr()
```

```python
# Create a correlation matrix using pandas (Merged)

subM.corr()
```

### Creating a Correlation Matrix Heatmap

```python
# Create a correlation heatmap using matplotlib

plt.matshow(sub22.corr())
plt.show()
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub22.corr(), annot=True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
plt.title('WHR Correlation Matrix Heatmap (2022)', fontsize=14) # add title
plt.show()
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(subM.corr(), annot=True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
plt.title('WHR Correlation Matrix Heatmap (Merged)', fontsize=14) # add title
plt.show()
```

**Correlation Results Findings** <br/>
Both the 2022 World Happiness Report and the Merged report paint a similar picture when it comes to correlation. GDP per Capita, Social Support, and Healthy Life Expectancy hold a fairly strong correlation to happiness score. Our Strongest correlation was between GDP per Capita and Healthy Life Expectancy (which makes sense as having more money for healthcare/quality of life will improve average lifespan). I think its worth looking at
the strongest correlations for each major category. It's also important to note that we should expect higher correlation values for the 2022 dataset vs. the merged one since we have significantly less rows.

## Creating Scatterplots

```python
# Create a scatterplot for the "GDP per Capita" and "Healthy Life Expectancy" columns in seaborn

sns.lmplot(x = 'GDP per Capita', y = 'Healthy Life Expectancy', data = sub22)
sns.lmplot(x = 'GDP per Capita', y = 'Healthy Life Expectancy', data = subM)
plt.show()
```

```python
# Create a scatterplot for the "Happiness Score" and "Healthy Life Expectancy" columns in seaborn

sns.lmplot(x = 'Happiness Score', y = 'Healthy Life Expectancy', data = sub22)
sns.lmplot(x = 'Happiness Score', y = 'Healthy Life Expectancy', data = subM)
plt.show()
```

```python
# Create a scatterplot for the "Social Support" and "Happiness Score" columns in seaborn

sns.lmplot(x = 'Social Support', y = 'Happiness Score', data = sub22)
sns.lmplot(x = 'Social Support', y = 'Happiness Score', data = subM)
plt.show()
```

```python
# Create a scatterplot for the "Social Support" and "Healthy Life Expectancy" columns in seaborn

sns.lmplot(x = 'Social Support', y = 'Healthy Life Expectancy', data = sub22)
sns.lmplot(x = 'Social Support', y = 'Healthy Life Expectancy', data = subM)
plt.show()
```

```python
# Create a scatterplot for the "Perceptions of Corruption" and "Freedom to make life choices" columns in seaborn

sns.lmplot(x = 'Perceptions of Corruption', y = 'Freedom to make life choices', data = sub22)
sns.lmplot(x = 'Perceptions of Corruption', y = 'Freedom to make life choices', data = subM)
plt.show()
```

**Scatterplot Results Findings** <br/>
From each of the scatterplots we are able to see the differences in correlation. Also having the 2022 dataset, makes it a lot easier to see any correlations simply because there is less data to plot. Some of the stronger correlations (GPD per Capita and Healthy Life Expectancy + Social Support vs Happiness Score) were a bit easier to see as they had stronger correlations.

### Defining Low/Medium/High Happiness Scores

```python
sub22.loc[sub22['Happiness Score'] < 4, 'Happiness Rating'] = 'Low Happiness'
```

```python
sub22.loc[(sub22['Happiness Score'] >= 4) & (sub22['Happiness Score'] < 6), 'Happiness Rating'] = 'Average Happiness'
```

```python
sub22.loc[sub22['Happiness Score'] > 6, 'Happiness Rating'] = 'High Happiness'
```

```python
sub22['Happiness Rating'].value_counts(dropna = False)
```

Most countries have average happiness or high happiness scores
